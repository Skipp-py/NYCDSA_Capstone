```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
df = pd.read_csv('Zip_Zri_AllHomesPlusMultifamily.csv', dtype={'RegionName': str})
```

```python
targets = df
targets
```

```python
targets2 = pd.read_csv('Zip_ZORI_AllHomesPlusMultifamily_Smoothed.csv')
targets2
```

```python
cols_2010 = targets.filter(regex='2010')
cols_2011 = targets.filter(regex='2011')
cols_2012 = targets.filter(regex='2012')
cols_2013 = targets.filter(regex='2013')
cols_2014 = targets.filter(regex='2014')
cols_2015 = targets.filter(regex='2015')
cols_2016 = targets.filter(regex='2016')
cols_2017 = targets.filter(regex='2017')
cols_2018 = targets.filter(regex='2018')
cols_2019 = targets.filter(regex='2019')
cols_2020 = targets.filter(regex='2020')
```

```python
cols_2010['average'] = cols_2010.mean(axis=1)
cols_2011['average'] = cols_2011.mean(axis=1)
cols_2012['average'] = cols_2012.mean(axis=1)
cols_2013['average'] = cols_2013.mean(axis=1)
cols_2014['average'] = cols_2014.mean(axis=1)
cols_2015['average'] = cols_2015.mean(axis=1)
cols_2016['average'] = cols_2016.mean(axis=1)
cols_2017['average'] = cols_2017.mean(axis=1)
cols_2018['average'] = cols_2018.mean(axis=1)
cols_2019['average'] = cols_2019.mean(axis=1)
cols_2020['average'] = cols_2020.mean(axis=1)
```

```python
cols_2017
```

```python
year_avgs = pd.DataFrame({'geo_id':df.RegionName,
             '2010':cols_2010.average,
             '2011':cols_2011.average,
             '2012':cols_2012.average,
             '2013':cols_2013.average,
             '2014':cols_2014.average,
             '2015':cols_2015.average,
             '2016':cols_2016.average,
             '2017':cols_2017.average,
             '2018':cols_2018.average,
             '2019':cols_2019.average,
             '2020':cols_2020.average})
year_avgs.head(60)
```

```python
year_avgs.to_csv('year_avgs.csv')
```

```python
df_2011 = pd.read_csv('./Data/demo_zip_11.csv', dtype={'geo_id': str})
df_2012 = pd.read_csv('./Data/demo_zip_12.csv', dtype={'geo_id': str})
df_2013 = pd.read_csv('./Data/demo_zip_13.csv', dtype={'geo_id': str})
df_2014 = pd.read_csv('./Data/demo_zip_14.csv', dtype={'geo_id': str})
df_2015 = pd.read_csv('./Data/demo_zip_15.csv', dtype={'geo_id': str}).drop(columns = 'geoid') #mysterious column of all NaN
df_2016 = pd.read_csv('./Data/demo_zip_16.csv', dtype={'geo_id': str})
df_2017 = pd.read_csv('./Data/demo_zip_17.csv', dtype={'geo_id': str})
df_2018 = pd.read_csv('./Data/demo_zip_18.csv', dtype={'geo_id': str})
```

```python
df_2011 = pd.merge(df_2011, year_avgs[['geo_id', '2011']], on='geo_id').rename(columns={'2011':'avg_rent'})
df_2012 = pd.merge(df_2012, year_avgs[['geo_id', '2012']], on='geo_id').rename(columns={'2012':'avg_rent'})
df_2013 = pd.merge(df_2013, year_avgs[['geo_id', '2013']], on='geo_id').rename(columns={'2013':'avg_rent'})
df_2014 = pd.merge(df_2014, year_avgs[['geo_id', '2014']], on='geo_id').rename(columns={'2014':'avg_rent'})
df_2015 = pd.merge(df_2015, year_avgs[['geo_id', '2015']], on='geo_id').rename(columns={'2015':'avg_rent'})
df_2016 = pd.merge(df_2016, year_avgs[['geo_id', '2016']], on='geo_id').rename(columns={'2016':'avg_rent'})
df_2017 = pd.merge(df_2017, year_avgs[['geo_id', '2017']], on='geo_id').rename(columns={'2017':'avg_rent'})
df_2018 = pd.merge(df_2018, year_avgs[['geo_id', '2018']], on='geo_id').rename(columns={'2018':'avg_rent'})
```

```python
df_2011.to_csv('./Data_w_targets/demo_zip_11.csv')
df_2012.to_csv('./Data_w_targets/demo_zip_12.csv')
df_2013.to_csv('./Data_w_targets/demo_zip_13.csv')
df_2014.to_csv('./Data_w_targets/demo_zip_14.csv')
df_2015.to_csv('./Data_w_targets/demo_zip_15.csv')
df_2016.to_csv('./Data_w_targets/demo_zip_16.csv')
df_2017.to_csv('./Data_w_targets/demo_zip_17.csv')
df_2018.to_csv('./Data_w_targets/demo_zip_18.csv')
```
