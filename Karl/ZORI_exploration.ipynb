```python
import pandas as pd
df = pd.read_csv('Metro_ZORI_AllHomesPlusMultifamily_SSA.csv')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
df.shape
```

```python
df.columns
```

```python
df[df['2021-10'] == df['2021-10'].min()].RegionName
```

```python
us_prices = df[df.RegionName == 'United States'].drop(columns=[
    'RegionID', 'RegionName', 'SizeRank']).T.reset_index().rename(columns={'index':'date', 0:'price'})
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(us_prices.date,us_prices.price)
plt.title('Rent prices in the US vs. time based on ZORI')
plt.xlabel('Date')
plt.ylabel('Rent Prices')
plt.show()
```
