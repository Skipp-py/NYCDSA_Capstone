```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pca = PCA(n_components = 70)
pipeline = Pipeline(
    [("scaling", StandardScaler()),('pca', pca)]
)
pipeline.fit(X,y)
```

```python
pca_norm = pipeline.named_steps["pca"]
pca_norm.components_
```

```python
pca.set_params(n_components = None)
pca.fit(X)
plt.plot(range(70), pca_norm.explained_variance_ratio_[:70])
plt.scatter(range(70), pca_norm.explained_variance_ratio_[:70])
plt.xlabel('ith components')
plt.ylabel('Percentage of Variance')
plt.show()
```

```python
pca_norm.set_params(n_components = 70)
pca_norm = pca_norm.fit_transform(X,y)
```

```python
inertia_list = []
for num_clusters in range(1,71):
    kmeans_model = KMeans(n_clusters=num_clusters, init = "k-means++")
    kmeans_model.fit(pca_norm)
    inertia_list.append(kmeans_model.inertia_)
```

```python
plt.plot(range(1,71), inertia_list)
plt.scatter(range(1,71), inertia_list)
```
